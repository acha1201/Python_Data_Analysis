# 项目：评估和清理英国电商公司销售数据

## 分析目标
此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。  
本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 数据说明

这是一个跨国的数据集，包含了2010年12月1日至2011年12月9日期间发生的所有交易，是一家总部设在英国并注册的非商店在线零售。该公司主要销售独特的全场合礼品。该公司的许多客户是批发商。  

| 字段          | 说明    |
|-------------|---------------|
| InvoiceNo   | 发票号码：6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。  |
| StockCode   | 产品代码：5位数，作为产品的唯一标识符。  |
| Description | 描述：产品名称。    |
| Quantity    | 数量：产品在交易中的数量。    |
| InvoiceDate | 发票日期和时间：交易发生的日期和时间。  |
| UnitPrice   | 单位价格：价格为英镑。  |
| CustomerID  | 客户ID：5位数，作为客户的唯一标识符。  |
| Country     | 国家：客户所居住的国家的名称。    |

## 读取数据
评估和清理电商销售数据

导入数据分析所需要的库，并通过 Pandas 的 `read_csv` 函数，将原始数据文件 "e-commerce1.csv" 里的数据内容，解析为 DataFrame，并赋值给变量 `original_data`。

```python
import pandas as pd
```

```python
original_data = pd.read_csv("./e-commerce1.csv")
```

```python
original_data.head()
```

## 评估数据
在这一部分，将对上一部分建立的 `original_data` 这个 DataFrame 所包含的数据进行评估。  
评估主要从两个方面进行：结构和内容，即整齐度和干净度。  
数据的结构性问题指不符合“每个变量为一列，每个观察值为一行，每个单元格是一个值”这三个标准；  
数据的内容性问题包括存在丢失数据、重复数据、不一致数据、无效数据和错误数据等。

```python
original_data.sample(10)
```

**步骤描述**  
从上述抽样的10条数据来看，数据符合“每个变量为一列，每个观察值为一行，每个单元格是一个值”；  
具体来看，每行是关于某商品的一次交易，每列是交易相关的各个变量，因此不存在结构性问题。

### 评估数据干净度

```python
original_data.info()
```

**步骤描述**   
从输出结果来看，数据共有541909条观察值，  
但变量 `Description` 和 `CustomerID` 存在缺失值，需要对其进行进一步分析；  
此外，`InvoiceDate` 的数据类型应为日期，`CustomerID` 的数据类型应为字符串，因此需要进行数据类型的转换。

#### 评估缺失数据

##### 进一步分析`Description`

```python
original_data[original_data["Description"].isnull()]
```

**步骤描述**  
从输出结果来看，`Description` 缺失的观察值，其对应的 `UnitPrice` 为0，  
所以，需要进一步验证是否所有 `Description` 缺失的观察值，其对应的 `UnitPrice` 都为0。

```python
original_data[(original_data["Description"].isnull()) & (original_data["UnitPrice"] != 0)]
```

**步骤描述**  
从输出结果来看，所有 `Description` 缺失的观察值，其对应的 `UnitPrice` 都为0（不存在不为0的观察值）。  
因本项目的目的是需要“根据市场销售数据，挖掘畅销产品，并制定提升营收的策略”，  
因此，`Description` 产品名称和 `UnitPrice` 价格均为0的数据，无法提供相应的参考价值，后续可以考虑将其删除。

##### 进一步分析`CustomerID`

```python
original_data[original_data["CustomerID"].isnull()]
```

**步骤描述**  
因 `CustomerID` 缺失的观察值，所对应的其他字段信息不完全缺失，所以不影响分析目标，所以可以不做处理。

#### 评估重复数据

在数据字段中，`InvoiceNo`、`StockCode` 以及 `CustomerID` 均为唯一标识， 
但同一张发票，可以对应多个商品和客户，同一个客户可以多次购买同一件商品，不同的订单中也可以包含同样的商品，  
所以，在本项目中，允许存在重复数据。

#### 评估不一致数据

在本项目中，可以进行评估的不一致数据可能存在于 `Country` 字段中，  
因此，需要进一步评估该字段是否存在不同值指代同一国家的情况。

```python
original_data["Country"].value_counts()
```

**步骤描述**  
根据上述结果，不存在不一致的数据。

#### 评估无效或错误数据
可以通过DataFrame的 `describe` 方法，对数值的统计信息进行快速了解。

```python
original_data.describe()
```

**步骤说明**  
从上述结果可以看出，`Quantity` 数量和 `UnitPrice`价格都存在负数，  
正常情况下，这两个值不应该为负，  
因此，需要进一步分析。  
另，`CustomerID` 应该是字符串类型，不应该被进行数值计算。

```python
original_data[original_data["Quantity"] < 0]
```

**步骤描述**  
根据上述结果，`Quantity`为0的观察值，其 `InvoiceNo` 均以“C”开头（表示该订单被退回）， 
为此，需要进一步判断是否所有 `Quantity`为0的观察值，其 `InvoiceNo` 均以“C”开头。

```python
original_data[(original_data["Quantity"] < 0) & (original_data["InvoiceNo"].str[0] != "C")]
```

```python
print(type(original_data["InvoiceNo"]))
```

**步骤说明**  
根据上述结果可以看出，存在 `Quantity` 小于0但 `InvoiceNo` 不以“C”开头的观察值，  
但上述结果中，其对应的 `UnitPrice` 为0， 
因此，继续进行筛选，查看是否在上述结果的基础上，其 `UnitPrice` 均为0。

```python
original_data[(original_data["Quantity"] < 0) & (original_data["InvoiceNo"].str[0] != "C") & (original_data["UnitPrice"] != 0)]
```

**步骤说明**  
因此，当交易数量 `Quantity` 为负数时，要么商品被退回，要么商品单价为0，  
所以，基于本项目的分析目标，这些数据不是有效成交数据，还会影响有效的数据分析，所以上述观察值需要进行删除。

```python
original_data[original_data["UnitPrice"] < 0]
```

**步骤说明**  
`UnitPrice` 为负的两条数据均为“Adjust bad deb”坏账调整，不是分析目标需要关注的内容，所以后续可以直接删掉。

## 清理数据

根据前面评估部分得到的结论，需要进行的数据清理包括： 
- 把 `InvoiceDate` 变量的数据类型转换为日期时间；
- 把 `CustomerID` 变量的数据类型转换为字符串；
- 把 `Description` 变量缺失的观察值删除；
- 把 `Quantity` 变量值为负数的观察值删除；
- 把 `UnitPrice` 变量值为负数的观察值删除。

为了区分经过清理的数据和原始的数据，创建新的变量 `cleaned_data` ，让它为 `original_data` 复制出的副本。  
之后的清理步骤都将在 `cleaned_data` 上进行。

```python
cleaned_data = original_data.copy()
cleaned_data.head()
```

把 `InvoiceDate` 变量的数据类型转换为日期时间类型⬇️

```python
cleaned_data["InvoiceDate"] = pd.to_datetime(cleaned_data["InvoiceDate"])
cleaned_data["InvoiceDate"]
```

把 `CustomerID` 变量的数据类型转换为字符串⬇️

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].astype("str")
cleaned_data["CustomerID"]
```

把 `CustomerID` 变量值结尾的 “.0” 删除。

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].str.slice(0, -2)
cleaned_data["CustomerID"]
```

把 `Description` 变量缺失的观察值删除⬇️  
用 `.dropna()` 可以直接删除含缺失值的观察值（整行），  
`.dropna(subset = ["列名"])` 则是删除该列中，是缺失值的观察值（行）。

```python
cleaned_data = cleaned_data.dropna(subset = ["Description"])
```

```python
cleaned_data["Description"].isnull().sum()
```

⬆️ 验证 `Description` 列中是否所有是缺失值的观察值都被删除  
`.isnull()` 返回是否是缺失值的布尔值，`False` or `True`，然后调用 `sum()` 对布尔值求和，其中，`False` 是0，`True` 是1。

**如果原表中有数据不一致的值，可采用以下步骤**  
把 `Country` 变量值 `"USA"` 替换为 `"United States"`，并检查替换后的 `"USA"` 变量个数。

```python
cleaned_data["Country"] = cleaned_data["Country"].replace({"USA": "United States"})
```

```python
len(cleaned_data[cleaned_data["Country"] == "USA"])
```

```python
len(cleaned_data[cleaned_data["Country"] == "United States"])
```

把 `Quantity` 变量值为负数的观察值删除，并验证⬇️

```python
cleaned_data = cleaned_data[cleaned_data["Quantity"] >= 0]
```

```python
len(cleaned_data[cleaned_data["Quantity"] < 0])
```

把 `UnitPrice` 变量值为负数的观察值删除，并验证⬇️

```python
cleaned_data = cleaned_data[cleaned_data["UnitPrice"] >= 0]
```

```python
len(cleaned_data[cleaned_data["UnitPrice"] < 0])
```
